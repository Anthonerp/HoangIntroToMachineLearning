```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# Load data
iris = load_iris()
X, y = iris.data, iris.target

# Compute CV accuracy for k=1..31 using cv=10
k_range = range(1, 32)
mean_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10)
    mean_scores.append(scores.mean())

mean_scores = np.array(mean_scores)

best_k = list(k_range)[mean_scores.argmax()]
best_acc = mean_scores.max()
print("Best k:", best_k, "Best CV accuracy:", best_acc)

# Plot 1: k=1..31
plt.figure(figsize=(8, 4))
plt.plot(list(k_range), mean_scores)
plt.xlabel("Value of K for KNN")
plt.ylabel("Cross-Validated Accuracy")
plt.xlim(0, 31)
plt.ylim(0.952, 0.981)
plt.show()

# Plot 2: zoom in to k=11..15
k_zoom = list(range(11, 16))
zoom_scores = mean_scores[10:15]

plt.figure(figsize=(8, 4))
plt.plot(k_zoom, zoom_scores)
plt.xlabel("Value of K for KNN")
plt.ylabel("Cross-Validated Accuracy")
plt.xlim(11, 15)
plt.ylim(0.966, 0.981)
plt.show()
```
